```python
df = pd.read_csv('menu.csv')
```

```python
df_melt = df.drop(columns=['gmail', 'github', 'domain', 'vercel', 'role'])
```

```python
df_melt.head(10)
```

```python
df_melt = df_melt.melt(id_vars=['ename'], var_name='dt', value_name='menu_name')
```

```python
df_melt.head(6)
```

```python
print(df_melt)
```

```python
df_2 = df_melt
df_2 = df_2[~df_2['menu_name'].isin(['<결석>', '-', 'x'])]
df_2 = df_2.dropna()
print(df_2['ename'].value_counts())
```

```python
df_3 = df_2.groupby('ename').size().sort_values(ascending=False)
print(df_3)
```
